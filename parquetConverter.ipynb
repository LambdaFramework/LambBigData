## Local Boolean Scheme conversion 

```python
import awkward
from array import array
import numpy as np

DFlb=[]
for element in localboolscheme:
    event=[]
    for jaggedElement in HWp.arrays("%s" %element)["%s" %element]:
        eventlet=[]
        for num,eve in enumerate(jaggedElement):
            eventlet.append(1*eve)
        event.append(eventlet)
        
    DFlb.append(pd.DataFrame(pd.Series([x for x in awkward.JaggedArray.fromiter(event)], name='%s' %element)))
```
